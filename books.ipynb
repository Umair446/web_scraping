```python
import requests
import bs4
import lxml
```

```python
base_url = 'http://books.toscrape.com/catalogue/page-{}.html'
```

```python
data = []
for i in range(1,51):
    print(f"Working on {i} page")    
    res = requests.get(base_url.format(f'{i}'))
    soup = bs4.BeautifulSoup(res.text,"lxml")
    products = soup.select(".product_pod")
    for prod in products:
        product_details = {}
        title =prod.select("h3")[0].select('a')[0].get('title').strip()
        link =prod.select("h3")[0].select('a')[0].get('href').strip()
        rating = prod.select(".star-rating")[0].get('class')[-1]
        price = prod.select(".price_color")[0].text.strip()
        
        availability_text = prod.select(".instock.availability")[0].text.strip()
        
        availability = "in stock" in availability_text.lower()
        product_details['title'] = title
        product_details['link']= link
        product_details['rating'] = rating
        product_details['price'] = price
        product_details['availability'] = availability  
        data.append(product_details)
        print(rating, price, availability,title,link)
```
